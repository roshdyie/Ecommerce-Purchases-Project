## How many Customers have a credit card that Expires in 2021?

```python
data.columns
```

```python
data['CC Exp Date']
```

### We need to split our data 

```python
def fun():
    count=0
    for date in data['CC Exp Date']:
        if date.split('/')[1]=='21':
            count=count+1
    print(count)
```

```python
fun()
```

### Another Method, using Lambda function 'Slicing'

```python
data['CC Exp Date'].apply(lambda x:x[3:]=='21')
```

```python
data[data['CC Exp Date'].apply(lambda x:x[3:]=='21')]
```

```python
len(data[data['CC Exp Date'].apply(lambda x:x[3:]=='21')])
```
